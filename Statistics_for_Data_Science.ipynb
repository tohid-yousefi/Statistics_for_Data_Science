### Hipotez Testinin UygulanmasÄ±

H0: Ã–rnek daÄŸÄ±lÄ±mÄ± ile teorik normal daÄŸÄ±lÄ±m arasÄ±nda istatistiksel olarak anlamlÄ± bir fark yoktur

H1: Ã–rnek daÄŸÄ±lÄ±mÄ± ile teorik normal daÄŸÄ±lÄ±m arasÄ±nda istatistiksel olarak anlamlÄ± bir fark vardÄ±r

```python
from scipy.stats import shapiro,ttest_1samp
from statsmodels.stats.descriptivestats import sign_test
T_istatistigi , p_value = shapiro(olcumler)
print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
if p_value < 0.05:
  print("Veriler Normal DaÄŸÄ±lmamÄ±ÅŸtÄ±r ve Nonparametrik Tek Ã–rneklem T Testi UygulanmalÄ±dÄ±r.")
  T_istatistigi , p_value = sign_test(olcumler, 170)
  if p_value < 0.05:
    print("Ã–rnek daÄŸÄ±lÄ±mÄ± ile teorik normal daÄŸÄ±lÄ±m arasÄ±nda istatistiksel olarak anlamlÄ± bir fark vardÄ±r. H0 hipotezi reddedilir.")
  else:
    print("Ã–rnek daÄŸÄ±lÄ±mÄ± ile teorik normal daÄŸÄ±lÄ±m arasÄ±nda istatistiksel olarak anlamlÄ± bir fark yoktur. H0 hipotezi reddedilemez.")
else:
  print("Veriler Normal DaÄŸÄ±lmÄ±ÅŸtÄ±r. Tek Ã–rneklem Testi Uygulanabilir.")
  T_istatistigi , p_value = stats.ttest_1samp(olcumler, popmean=170)
  if p_value < 0.05:
    print("Ã–rnek daÄŸÄ±lÄ±mÄ± ile teorik normal daÄŸÄ±lÄ±m arasÄ±nda istatistiksel olarak anlamlÄ± bir fark vardÄ±r. H0 hipotezi reddedilir.")
  else:
    print("Ã–rnek daÄŸÄ±lÄ±mÄ± ile teorik normal daÄŸÄ±lÄ±m arasÄ±nda istatistiksel olarak anlamlÄ± bir fark yoktur. H0 hipotezi reddedilemez.")
```

# Tek Ã–rneklem Oran Testi

ðŸ“Œ Oransal bir ifade test edilmek istenildiÄŸinde kullanÄ±lÄ±r.

H0: p=0.125

H1: p!=0.125

```python
from statsmodels.stats.proportion import proportions_ztest
count = 40
nobs = 500
value = 0.125
T_istatistigi, p_value = proportions_ztest(count, nobs, value)
print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
if p_value < 0.05:
  print("H0 hipotezi reddedilir.")
else:
  print("H0 hipotezi reddedilemez.")
```

# Ä°ki Ã–rneklem Oran Testi

H0: FarklÄ±lÄ±k yoktur.

H1: FarklÄ±lÄ±k vardÄ±r.

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

basari_sayisi = np.array([300,250])
gozlem_sayisi = np.array([1000,1100])
T_istatistigi, p_value = proportions_ztest(count=basari_sayisi, nobs=gozlem_sayisi)
print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
if p_value < 0.05:
  print("H0 hipotezi reddedilir.")
else:
  print("H0 hipotezi reddedilemez.")
```

# BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem T Testi(AB Testi)

ðŸ“Œ Ä°ki grup ortalamasÄ± arasÄ±nda karÅŸÄ±laÅŸtÄ±rma yapÄ±lmak istenildiÄŸinde kullanÄ±lÄ±r.

H0: M1 = M2 

H1: M1 != M2

```python
# Veri Tipi I
import pandas as pd
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,
        25,21,18,24,26,33,26,27,28,19,25])

B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,
        32,33,33,33,31,32,33,26,32,33,29])
A_B = pd.concat([A,B],axis=1)
A_B.columns = ["A","B"]
A_B.head()
```

```python
# Veri Tipi II
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,
        25,21,18,24,26,33,26,27,28,19,25])

B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,
        32,33,33,33,31,32,33,26,32,33,29])

# A ve A'nÄ±n grubu
Group_A = pd.DataFrame(np.arange(len(A)))
Group_A[:] = "A"
A = pd.concat([A,Group_A], axis = 1)

# B ve B'nin Grubu
Group_B = pd.DataFrame(np.arange(len(B)))
Group_B[:] = "B"
B = pd.concat([B,Group_B], axis = 1)

# TÃ¼m Veri
AB = pd.concat([A,B])
AB.columns = ["gelir","Grup"]
AB.head()
```

```python
import seaborn as sns
sns.boxplot(x="Grup", y="gelir", data=AB);
```

## VarsayÄ±mlar

1) Normallik

2) Varyans HomojenliÄŸi


```python
# normallik varsayÄ±mÄ±
# Shapiro-Wilks Test
from scipy.stats import shapiro
A_T_istatistigi , A_p_value = shapiro(A_B.A)
B_T_istatistigi , B_p_value = shapiro(A_B.B)
print("A iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(A_T_istatistigi, A_p_value))
print("B iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(B_T_istatistigi, B_p_value))
if A_p_value < 0.05 or B_p_value < 0.05:
  print("Veriler Normal DaÄŸÄ±lmamÄ±ÅŸtÄ±r.")
else:
  print("Veriler Normal DaÄŸÄ±lmÄ±ÅŸtÄ±r.")
```

```python
# varyans homojenliÄŸi
# levene testi
import scipy.stats as stats
T_istatistigi, p_value = stats.levene(A_B.A,A_B.B)
print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
if p_value < 0.05 :
  print("Varyanslar Homojen DeÄŸildir")
else:
  print("Varyanslar Homojendir")
```

### Hipotez Testinin UygulanmasÄ±

H0: FarklÄ±lÄ±k yoktur.

H1: FarklÄ±lÄ±k vardÄ±r.

```python
from scipy.stats import shapiro
import scipy.stats as stats
A_T_istatistigi , A_p_value = shapiro(A_B.A)
B_T_istatistigi , B_p_value = shapiro(A_B.B)
print("A iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(A_T_istatistigi, A_p_value))
print("B iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(B_T_istatistigi, B_p_value))
if A_p_value < 0.05 or B_p_value < 0.05:
  print("Veriler Normal DaÄŸÄ±lmamÄ±ÅŸtÄ±r.")
else:
  print("Veriler Normal DaÄŸÄ±lmÄ±ÅŸtÄ±r.")
  T_istatistigi, p_value = stats.levene(A_B.A,A_B.B)
  print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
  if p_value < 0.05 :
    print("Varyanslar Homojen DeÄŸildir")
  else:
    print("Varyanslar Homojendir")
    T_istatistigi , p_value = stats.ttest_ind(A_B["A"],A_B["B"], equal_var=True)
    print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
    if p_value < 0.05:
      print("FarklÄ±lÄ±k vardÄ±r. H0 hipotezi reddedilir.")
    else:
      print("FarklÄ±lÄ±k yoktur. H0 hiptezi reddedilemez.")

```

# Nonparametrik BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem T Testi

ðŸ“Œ BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem T Testini uyguladÄ±ÄŸÄ±mÄ±za eÄŸer varsayÄ±mlar saÄŸlanmadÄ±ysa Nonparametrik BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem T Testi uygulanÄ±r.

```python
# Veri Tipi I
import pandas as pd
import numpy as np
A = pd.DataFrame(pd.DataFrame(np.random.normal(0,200,50)))
B = pd.DataFrame(pd.DataFrame(np.random.randint(0,200,50)))
A_B = pd.concat([A,B],axis=1)
A_B.columns = ["A","B"]
A_B.head()
```

```python
# Veri Tipi II
# A ve A'nÄ±n grubu
Group_A = pd.DataFrame(np.arange(len(A)))
Group_A[:] = "A"
A = pd.concat([A,Group_A], axis = 1)

# B ve B'nin Grubu
Group_B = pd.DataFrame(np.arange(len(B)))
Group_B[:] = "B"
B = pd.concat([B,Group_B], axis = 1)

# TÃ¼m Veri
AB = pd.concat([A,B])
AB.columns = ["gelir","Grup"]
AB.head()
```

```python
import seaborn as sns
sns.boxplot(x="Grup", y="gelir", data=AB);
```

## VarsayÄ±mlar

1) Normallik Testi SaÄŸlanmamalÄ±

2) Varyanslar Homojen OlmamalÄ±

**Not**: YukarÄ±daki ÅŸartlar projeye gÃ¶re hafifletilebilir.

```python
# normallik varsayÄ±mÄ±
# Shapiro-Wilks Test
from scipy.stats import shapiro
A_T_istatistigi , A_p_value = shapiro(A_B.A)
B_T_istatistigi , B_p_value = shapiro(A_B.B)
print("A iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(A_T_istatistigi, A_p_value))
print("B iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(B_T_istatistigi, B_p_value))
if (A_p_value < 0.05) or (B_p_value < 0.05):
  print("Veriler Normal DaÄŸÄ±lmamÄ±ÅŸtÄ±r.")
else:
  print("Veriler Normal DaÄŸÄ±lmÄ±ÅŸtÄ±r.")
```

```python
# varyans homojenliÄŸi
# levene testi
import scipy.stats as stats
T_istatistigi, p_value = stats.levene(A_B.A,A_B.B)
print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
if p_value < 0.05 :
  print("Varyanslar Homojen DeÄŸildir")
else:
  print("Varyanslar Homojendir")
```

# BaÄŸÄ±mlÄ± Ä°ki Ã–rneklem T Testi(AB Testi)

ðŸ“Œ BaÄŸÄ±mlÄ± Ä°ki grup ortalamasÄ± arasÄ±nda karÅŸÄ±laÅŸtÄ±rma yapÄ±lmak istenildiÄŸinde kullanÄ±lÄ±r.

H0: M1 = M2 

H1: M1 != M2

```python
# Veri Tipi I
import numpy as np
import pandas as pd
oncesi = pd.DataFrame([123,119,119,116,123,123,121,120,117,118,121,121,123,119,
            121,118,124,121,125,115,115,119,118,121,117,117,120,120,
            121,117,118,117,123,118,124,121,115,118,125,115])

sonrasi = pd.DataFrame([118,127,122,132,129,123,129,132,128,130,128,138,140,130,
             134,134,124,140,134,129,129,138,134,124,122,126,133,127,
             130,130,130,132,117,130,125,129,133,120,127,123])
A_B = pd.concat([oncesi,sonrasi],axis=1)
A_B.columns = ["Oncesi","Sonrasi"]
A_B.head()
```

```python
# Veri Tipi II
oncesi = pd.DataFrame([123,119,119,116,123,123,121,120,117,118,121,121,123,119,
            121,118,124,121,125,115,115,119,118,121,117,117,120,120,
            121,117,118,117,123,118,124,121,115,118,125,115])

sonrasi = pd.DataFrame([118,127,122,132,129,123,129,132,128,130,128,138,140,130,
             134,134,124,140,134,129,129,138,134,124,122,126,133,127,
             130,130,130,132,117,130,125,129,133,120,127,123])

# A ve A'nÄ±n grubu
Grup_Oncesi = pd.DataFrame(np.arange(len(oncesi)))
Grup_Oncesi[:] = "ONCESI"
A = pd.concat([oncesi,Grup_Oncesi],axis=1)

# B ve B'nin Grubu
Grup_Sonrasi = pd.DataFrame(np.arange(len(sonrasi)))
Grup_Sonrasi[:] = "SONRASI"
B = pd.concat([sonrasi,Grup_Sonrasi],axis=1)

# TÃ¼m Veri
AB = pd.concat([A,B])
AB.columns = ["PERFORMANS","ONCESI_SONRASI"]
AB.head()
```

```python
import seaborn as sns
sns.boxplot(x="ONCESI_SONRASI", y="PERFORMANS", data=AB);
```

## VarsayÄ±mlar

1) Normallik

2) Varyans HomojenliÄŸi

```python
# normallik varsayÄ±mÄ±
# Shapiro-Wilks Test
from scipy.stats import shapiro
A_T_istatistigi , A_p_value = shapiro(A_B.Oncesi)
B_T_istatistigi , B_p_value = shapiro(A_B.Sonrasi)
print("Oncesi iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(A_T_istatistigi, A_p_value))
print("Sonrasi iÃ§in: T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(B_T_istatistigi, B_p_value))
if (A_p_value < 0.05) or (B_p_value < 0.05):
  print("Veriler Normal DaÄŸÄ±lmamÄ±ÅŸtÄ±r.")
else:
  print("Veriler Normal DaÄŸÄ±lmÄ±ÅŸtÄ±r.")
```

```python
# varyans homojenliÄŸi
# levene testi
import scipy.stats as stats
T_istatistigi, p_value = stats.levene(A_B.Oncesi,A_B.Sonrasi)
print("T istatistiÄŸi: %.4f , P value deÄŸeri: %.4f" %(T_istatistigi, p_value))
if p_value < 0.05 :
  print("Varyanslar Homojen DeÄŸildir")
else:
  print("Varyanslar Homojendir")
```

### Hipotez Testinin UygulanmasÄ±

H0: FarklÄ±lÄ±k yoktur.
